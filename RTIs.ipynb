```python
import glob
xml_files = glob.glob("*.xml")
xml_files.sort()
xml_files
```

```python
FP_list = ['AtomPairs2DCount',
           'AtomPairs2D',
           'EState',
           'CDKextended',
           'CDK',
           'CDKgraphonly',
           'KlekotaRothCount',
           'KlekotaRoth',
           'MACC',
           'Pubchem',
           'SubstructureCount',
           'Substructure']
```

```python
fp = dict(zip(FP_list, xml_files))
fp
```

```python
selection = ['canonical_smiles','Class']
df3_selection = df2[selection]
df3_selection.to_csv('molecule.smi', sep='\t', index=False, header=False)
```

```python
! cat molecule.smi | head -5
```

```python
from padelpy import padeldescriptor
fingerprint = 'Pubchem'
Pubchem_output_file = ''.join([fingerprint, '.excel']) #'.csv'
fingerprint_descriptortypes = fp[fingerprint]

padeldescriptor(mol_dir='molecule.smi',
                  d_file=Pubchem_output_file,#'AtomPairs2Dcount.csv'
                  #descriptortypes='AtomFingerprint.xml'),
                  descriptortypes= fingerprint_descriptortypes,
                  detectaromaticity=True,
                  standardizenitro=True,
                  standardizetautomers=True,
                  threads=2,
                  removesalt=True,
                  log=True,
                  fingerprints=True)
```

```python
Descriptors =pd.read_csv(Pubchem_output_file)
Descriptors
```

```python
Pubchem = Descriptors.drop(columns='Name')
Pubchem
```

```python
df_combined = pd.concat([Pubchem, df_class], axis=1)
df_combined
```

Exporting the Python Dataframe into a Comma separated Value (CSV) file for further analysis in RStudio.  

```python
df_combined.to_csv('RTIs.csv', index=False)
```
